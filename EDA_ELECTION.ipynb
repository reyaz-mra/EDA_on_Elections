```python
winner_member = df[df['WINNER'] ==1 ]
```

```python
winner_member = winner_member[['NAME','PARTY','TOTAL\nVOTES','TOTAL ELECTORS']]
```

```python
winner_member['VOTE_PERCENT'] = (100 * winner_member['TOTAL\nVOTES']/winner_member['TOTAL ELECTORS']).round()
```

```python
winner_member
```

```python
all_top_winners = winner_member.sort_values("VOTE_PERCENT",ascending=False)
```
